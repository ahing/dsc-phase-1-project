```python
# Import standard packages
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
```

```python
# Here you run your code to explore the data
import glob, os #file search tool
fpath = 'zippedData/'
os.listdir(fpath) #same as typing ls in terminal
```

```python
query = fpath+ '*.gz'
```

```python
file_list = glob.glob(query)
#file_list
```

```python
#source for this code: https://youtube.com/watch?v=BKpSVE0VF0U
#This is necessary to learn what tables may be useful in analysis
tables = {}
for file in file_list:
    print('---'*20)
    file_name = file.replace('zippedData/', '').replace('.', '_')
    print(file_name)
    if '.tsv.gz' in file:
        temp_df = pd.read_csv(file, sep = "\t", encoding = "latin-1")
    else:
        temp_df = pd.read_csv(file)
    display(temp_df.head(), temp_df.tail())
    
    
    tables[file_name] = temp_df
```

```python
bom_movie_gross = tables['bom_movie_gross_csv_gz'].copy()
#bom_movie_gross.head()
```

```python
imdb_title_basics = tables['imdb_title_basics_csv_gz'].copy()
tmdb_movies = tables['tmdb_movies_csv_gz'].copy()
#tmdb_movies.head()
rt_movie_info = tables['rt_movie_info_tsv_gz'].copy()
#rt_movie_info.head()
```

```python
df1 = pd.merge(bom_movie_gross, imdb_title_basics, left_on = 'title', 
              right_on = 'primary_title')
#df1
```

```python
df1[df1.duplicated(keep=False, subset=['title', 'original_title','domestic_gross'])]
```

```python
#check that all duplicates removed
df1 = df1.drop_duplicates(keep='first', subset=['title', 'original_title','domestic_gross'])
df1[df1.duplicated(keep=False, subset=['title', 'original_title','domestic_gross'])]
```

```python
#Sorting by most made domestic
df1.sort_values(by='domestic_gross', ascending=False)
```

```python
#creating column for total gross
#remove NAs
df1['foreign_gross']= df1['foreign_gross'].fillna('0')
df1['foreign_gross'] = df1['foreign_gross'].map(lambda x: x.replace(',', ''))
df1['foreign_gross'] = pd.to_numeric(df1['foreign_gross']) #convert this column to float
#summing all gross to get worldwide amount
sum_gross = df1['domestic_gross'] + df1['foreign_gross']
#creating new column for sum
df1['total_gross'] = sum_gross
#df1.info() #type of each column domestic is float and foreign is object
#df1.head()
```

```python
#merge to find movie budget
df2 = pd.merge(df1, tmdb_movies, on = 'title', )
#Creating a DF with columns that I will need to answer questions
needed_columns = df2[['title', 'domestic_gross', 'foreign_gross',  
                      'popularity', 'release_date', 'vote_average', 'vote_count', 'genres', 'total_gross',
                     'studio']]
needed_columns.head()
```

After merging the two dataframes and previewing them, I noticed there was a lot of duplicates of the same movie. This portion was needed to remove those duplicates to have a move accurate final data table. There was a final check to make sure all duplicates were removed. Duplicates would skew the final financial result which would make it seem higher grossed than it actually is.

```python
needed_columns = needed_columns.dropna(subset = ['genres'])
```

```python
#ex = needed_columns.loc[0, 'genres']
#ex.split(',')
#Separating Genres and making them into a list in their own new column 
genres_split = needed_columns['genres'].map(lambda x: x.split(','))
needed_columns['genres_split'] = genres_split
needed_columns.head()
```

```python
#Create new DF with each type of genre in a new row
df_genres = needed_columns.explode('genres_split')
df_genres.head()
```

```python
#Sorting Genres from Highest Gross to least
genres_avg = df_genres.groupby('genres_split')['total_gross'].mean().sort_values(ascending=False)
genres_avg.index
```

```python
plt.figure(figsize = (10,4)) #10 inches wide 4 inches high to avoid x axis overlap
ax = sns.barplot(data = df_genres, x = 'genres_split', y = 'total_gross',
           order = genres_avg.index)
ax.set(title='Highest Grossed Genres')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45,ha='right');
plt.savefig(fname='./images/gross_genre.png', facecolor='white') #save in images folder
```

### Q: What genres are the most successful?  

After joining several tables, I was able to see some important information about each type of movie. I got it to the point where I could see title, total gross (domestic + foreign) and genre of each movie. Using this combined data and removing duplicates I could see from a bar plot that Adventure, Animation, Action, Sci-Fi, and Fantasy were among the top grossed genres with the least standard deviation. Given the above results, I would recommend  that Microsoft will see the most profits world wide creating these type of movies. After fine tuning the graph by sorting highest gross in descending order, my initial observation was close to accurate. The top 5 highest gross genres are: Adventure, Sci-Fi, Animation, Action, and Fantasy respectively.

```python
#sorting by total gross
df_tg = needed_columns.sort_values(by='total_gross', ascending=False)
#dropping duplicates, keeping first copy
df_tg = df_tg.drop_duplicates(keep='first', subset=['title'])
df_tg.head(10)
#df_tg.info() release date is an object
```

```python
#converting string release date to a type datetime
df_tg['release_date']  = pd.to_datetime(df_tg['release_date'])
df_tg.info() #check type change
```

```python
df_tg['release_month'] = df_tg['release_date'].dt.month_name() #could save as new column
```

```python
#Ordering 
movies_avg = df_tg.groupby('release_month')['total_gross'].mean().sort_values(
    ascending=False)
movies_avg.index
```

```python
plt.figure(figsize = (15,4)) #15inches wide 4 inches high to avoid x axis overlap
sns.barplot(data = df_tg, x = 'release_month', y = 'total_gross', order =movies_avg.index
           ).set(title='Best Time to Release a Movie')
plt.savefig(fname='./images/release_month.png', facecolor='white')
```

### Q: What time of year to release a movie?

Building off the previous dateframe created, total gross was sorted in descending order. From there, duplicates were removed a new column named "release month" was created. This new column contained a datetime version of the string "release date" column. From there, the new column would have the release month of every movie. After graphing this dataframe with release month and total gross as variables, I was able to see the how well movies did in every month. Well was determined to mean which month has the most gross value. With the created bar graph, it can provide Microsoft with the insight of which month may be best to release a film to get the most revenue. This graph was improved by sorting the highest grossed months in descending order. As a result, a suggestion to Microsoft will be to release movies in: July, June, May, December, or November. Further research can be done with a marketing team as to why these months are the most popular.

```python
#ordering highest grossed studio in descending order
studio_avg = df_tg.groupby('studio')['total_gross'].mean().sort_values(ascending=False)
studio_avg.index #checking that it worked
```

```python
plt.figure(figsize = (15,4)) #15inches wide 4 inches high to avoid x axis overlap
sns.barplot(data = df_tg, x = 'studio', y = 'total_gross', 
            order=studio_avg.index[:10]).set(title='Highest Grossing Studio')
plt.savefig(fname='./images/gross_studio.png', facecolor='white')
```

Doing similar sorting as the previous graph, the x axis was replaced with studio. This time there were too many studios in the table, so this was limited to the top ten. The highest grossed ones are: P/DW, BV, GrtIndia, Fox, and Sony respectively
